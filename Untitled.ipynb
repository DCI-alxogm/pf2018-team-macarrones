```python
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
x=np.random.rand(2) 
y=np.random.rand(2)
plt.plot(x,y,'p')
p=np.array([x,y]) 
print (p)
```

```python
for s in range (2):
    u=p[0,s]
    l=p[1,s]
    plt.scatter(u,l, color='pink') 
    for c in range(0,360,30):
            y=((0.1)*np.sin(c))
            x=((0.1)*np.cos(c))
            plt.scatter(x+u,y+l,color='blue')
```
